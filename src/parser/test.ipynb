```python
from utils import calculate_follow, calculate_first

terminals = {'|', 'c', 'a', ')', '(', '*', 'b'}
non_terminals = {'A', 'T', 'F', 'E'}

productions = {
    'E': [['E', '|', 'T'], ['T']],
    'T': [['T', 'F'], ['F']],
    'F': [['F', '*'], ['A']],
    'A': [['a'], ['b'], ['c'], ['(', 'E', ')']]
}

start_symbol = 'E'

wrong_non_terminals = {'E', 'F', 'T', 'A'}

first = calculate_first(productions, terminals, non_terminals)
second = calculate_first(productions, terminals, wrong_non_terminals)

for nt in non_terminals:
    try:
        assert first[nt] == second[nt]    
    except:
        print(first[nt])
        print(second[nt])

first = calculate_follow(productions, terminals, non_terminals, start_symbol) 
second = calculate_follow(productions, terminals, wrong_non_terminals, start_symbol)

for nt in non_terminals:
    try:
        assert first[nt] == second[nt]    
    except:
        print(first[nt])
        print(second[nt])
```
